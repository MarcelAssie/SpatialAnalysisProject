```python
import pandas as pd
from scipy import stats
```

```python
path = r"../../data/output/descriptive_analytic_data"
analytic_data_path = path + r"\analytic_data.csv"
descriptive_data_path = path + r"\descriptive_data.csv"
```

```python
analytic_data = pd.read_csv(analytic_data_path, delimiter=",")
descriptive_data = pd.read_csv(descriptive_data_path, delimiter=",")
```

```python
print('Moyenne des flux total par année:')
print(analytic_data.groupby('ANNEE')['TOTAL_FLUX'].mean())

print('Nombre de trajets par annee:')
print(analytic_data.groupby('ANNEE')['TRAJET'].count())

print('Moyenne des flux total par trajet:')
print(analytic_data.groupby('TRAJET')['TOTAL_FLUX'].mean())

print('Moyenne des flux total par trajet par annee:')
print(analytic_data.groupby(['ANNEE', 'TRAJET'])['TOTAL_FLUX'].mean())
```

```python
print("Analyse de l'impact de la DURÉE (DELTA_DUREE vs DELTA_PART_) ")
modes = ['MP', 'TC', 'VT', 'VL']
for mode in modes:
    col_y = f'DELTA_PART_{mode}'
    if col_y in analytic_data.columns:
        mask = analytic_data[['DELTA_DUREE', col_y]].notna().all(axis=1)
        x = analytic_data.loc[mask, 'DELTA_DUREE']
        y = analytic_data.loc[mask, col_y]
        
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
        print(f"Mode {mode}:")
        print(f"  R^2 = {r_value**2:.4f}")
        print(f"  Pearson r = {r_value:.4f}")
        print(f"  p-value = {p_value:.4e}")
        print(f"  Pente = {slope:.4f}")
```

```python
print("\n Analyse de la COMPETITION (DELTA_PART_TC vs autres) ")
other_modes = ['MP', 'VT', 'VL']
col_x = 'DELTA_PART_TC'
for mode in other_modes:
    col_y = f'DELTA_PART_{mode}'
    if col_x in analytic_data.columns and col_y in analytic_data.columns:
        mask = analytic_data[[col_x, col_y]].notna().all(axis=1)
        x = analytic_data.loc[mask, col_x]
        y = analytic_data.loc[mask, col_y]
        
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
        print(f"TC vs {mode}:")
        print(f"  R^2 = {r_value**2:.4f}")
        print(f"  Pearson r = {r_value:.4f}")
        print(f"  p-value = {p_value:.4e}")
        print(f"  Pente = {slope:.4f}")
```
